```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by 'City' and calculate mean of 'Bill'
cities_mean = df_bill.groupby('City')['Bill'].mean()

# Plot the histogram using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(cities_mean, bins=10, kde=True, color='blue', edgecolor='black')  # kde=False to disable the density curve
plt.title('Distribution of Mean Bill by City')
plt.xlabel('Mean Bill')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
