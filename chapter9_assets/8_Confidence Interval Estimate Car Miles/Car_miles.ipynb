## NOw with the help of confidence level we will find out the mean of the population

```python
sample_size = 100

sample = df_car.sample(sample_size, random_state=100)
sample.head()
```

```python
# mean_sample = sample.miles.mean()
# std_dev_sample = sample.miles.std()

# mean_sample, std_dev_sample

mean_sample = sample.miles.mean()
std_dev_sample = sample.miles.std()

mean_sample, std_dev_sample
```

```python
mean_population = df_car.miles.mean()
std_dev_population = df_car.miles.std()

mean_population, std_dev_population
```

```python
# Confidence Level |  Z-score
#  68.7%                1
#  95.5%                2
#  95%                  ?
```

```python
# for 95% CL we go left side from 2(appox == 1.9) Z-score because it is very close

# so after 95% the 2.5% is remaining in left and right hand side 

# so we will calculate for left side 95% + 2.5% == 97.5% it is commulative probability that the sample will lie in that range === Confidence Level

# probalility == 0.975 and from Z-score table the [Z-score == 1.96] === [import scipy.stats as st]
```

```python
import scipy.stats as st

z_score_95_confidence = st.norm.ppf(0.975)
z_score_95_confidence
```

```python
# margin_of_error = (Z-score)*(Standard Error)

# Standard Error = SD of population / sq. root of population(N)
```

```python
import numpy as np

margin_of_error = z_score_95_confidence * (std_dev_population/np.sqrt(sample_size))
margin_of_error
```

```python
# Confidence Interval = mean +&- standard error
```

```python
lower = mean_sample - margin_of_error
upper = mean_sample + margin_of_error

lower, upper
```
